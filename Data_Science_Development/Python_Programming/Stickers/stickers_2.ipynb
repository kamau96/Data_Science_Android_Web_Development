```python
# Data used for time series analysis.
path = "../../../Data/playground-series-s5e1/"
train = pd.read_csv(path+"train.csv", 
                    index_col=["date"], 
                    parse_dates=["date"])
test = pd.read_csv(path+"test.csv", 
                   index_col=["date"],
                   parse_dates=["date"])
```

```python
product_mean = train.groupby(["date", "country", "product"])["num_sold"].mean()
```

```python
for index, row in train.iterrows():
    if pd.isnull(row["num_sold"]):
        train[row["id"]] = product_mean[(index, row["country"], row["product"])]
```
